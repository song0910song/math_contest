```python
import torch
import matplotlib.pyplot as plt
```

```python
cluster = torch.ones(500, 2)
```

```python
data0 = torch.normal(4*cluster, 2)
data1 = torch.normal(-4*cluster, 2)
```

```python
label0 = torch.zeros(500)
label1 = torch.ones(500)
```

```python
x = torch.cat((data0, data1), dim=0)
y = torch.cat((label0, label1), dim=0)
```

```python
x,y
```

```python
plt.scatter(x.numpy()[:, 0], x.numpy()[:, 1], c=y.numpy(), cmap='coolwarm')
```
